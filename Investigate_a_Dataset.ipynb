```python
# Drop the unused columns from the Dataset
df.drop(['id','imdb_id','popularity','homepage','keywords','overview','production_companies','vote_count','vote_average','budget_adj','revenue_adj'] , axis = 1, inplace = True)
df.head()
```

```python
#check if there is any duplicated rows
df.duplicated().any()
```

```python
# Drop the duplicated rows but keep only the last duplicated row
df.drop_duplicates(keep='last' , inplace = True)
#again,check if there is any dplicated rows
df.duplicated().any()
```

```python
# Drop any null values
df.dropna(inplace = True)
df.info()
```

```python
df.head()
```

```python
#print number of rows with budget's value less than or equal to zero
df[df.budget<=0].shape[0]
```

```python
#Keep only the rows with budget's value greater than zero
df = df[df.budget>0]
```

```python
#Keep only the rows with revenue's value greater than zero
df = df[df.revenue>0]
##print number of rows with revenue's value less than or equal to zero
df[df.revenue<=0].shape[0]
```

```python
df.head()
```

```python
#Convert release date type from string to datetime 
df.release_date = pd.to_datetime(df.release_date)
df.info()
```

# Perform EDA

```python
df.describe()
```

```python
#function to find Maximum and minimum values in column
def max_ (column1):
  return(df.original_title[df[column1]==df[column1].max()],df[column1][df[column1]==df[column1]].max())
#function to find Minimum and minimum values in column
def min_(column1):
    return(df.original_title[df[column1]==df[column1].min()],df[column1][df[column1]==df[column1]].min())
```

# Research Question 1 (What is the the original title of the Movie with Maximum and Minimum budget?)

```python
#Movies with the Maximum and Minimum budget
print("The original title of the Movie with Maximum budget and its budget = {} \n ".format(max_("budget")))
print("The original title of the Movies with Minimum budget are and there budget = {} \n ".format(min_("budget")))
```

Now we known that the Movie with Maximum budget is"The Warrior's Way" and its budget = 425000000
Now we known that the Movie with Minimum budget are"Lost & Found" and "Love, Wedding, Marriag", and there budgets = 1

# Research Question 2  (What is the the original title of the Movie with Maximum and Minimum revenue?)

```python
#Movies with the Maximum and Minimum revenue	
print("The original title of the Movie with Maximum revenue and its revenue = {} \n ".format(max_("revenue")))
print("The original title of the Movies with Minimum revenue and there revenue = {} \n ".format(min_("revenue")))
```

Now we known that the Movie with Maximum revenue is"Avatar" and its revenue = 278150587 
Now we known that the Movie with Minimum revenue is "Mallrats" and there revenue = 2

# Research Question 3  (What is the the original title of the Movie with Maximum and Minimum profit?)

```python
#Calculate the profit of each Movie and insert it to third column
df.insert(2,'profit',df['revenue']-df['budget']) 
df.head()
```

```python
#Movies with Maximim and Minimum profit
print("The original title of the Movie with Maximum profit and its profit = {} \n ".format(max_("profit")))
print("The original title of the Movies with Minimum profit and there profit = {} \n ".format(min_("profit")))
```

Now we known that the Movie with Maximum profit is "Avatar" and its profit = 2544505847. We also known that the Movie with Minimum profit is "The Warrior's Way" and there profit = -413912431

# Research Question 5  (Which genre has the highest and lowest number of movies ?)

```python
def highest_lowest(column):
   no = pd.Series(df[column].str.cat(sep = '|').split('|')).value_counts(ascending = False)
   return (no)
```

```python
# concat all the data with | from the whole column then split it using | and count the number of times it occured. 
genre_=highest_lowest("genres")
genre_
```

```python
pt.figure(figsize=(18,20))

genre_.plot(kind='pie',fontsize =15)
# Show the plot
pt.show()
```

The genre who has the highest number of movies is "Drama" 
The genre who has the lowest number of movies "TV Movie"

# Research Question 6  (Which cast has the highest number of movies ?)

```python
# concat all the data with | from the whole column then split it using | and count the number of times it occured. 
cast_=highest_lowest("cast").head(30)
cast_
```

```python
pt.figure(figsize=(18,20))

cast_.plot(kind='pie',fontsize =15)
# Show the plot
pt.show()
```

The cast who has the highest number of movies is "Robert De Niro" 
Next after Robert De Niro is "Bruce Willis"

# Research Question 7(Is the number of movies increases or decrease over years?) 

```python
pt.figure(figsize=(15,17))
# count the number of movies in each year and sort it
no_movies = df['release_year'].value_counts().sort_index()
#plotting the result
no_movies.plot('bar',title='Number Of Movies vs Year')

pt.xlabel('Number Of Movies',fontsize=20)

pt.ylabel('Year')
pt.show()
```

We notice that the number of movies "increases" over years.

# Research Question 8 (Is the profit increases or decreases over years?)Â¶)

```python
#Group the Profit of each year then calculate the average
avg_profit = df.groupby('release_year')['profit'].mean()
#plotting
pt.figure(figsize=(12,10))
avg_profit.plot('line',title='Average Profit VS Years')

pt.xlabel('Year')

pt.ylabel('Profit')
```

The profit 'increases' over years
